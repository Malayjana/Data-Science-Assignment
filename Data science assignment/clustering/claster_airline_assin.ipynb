```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
airline = pd.read_csv('Downloads/Airlines.csv')
airline
```

```python
airline.info()
```

```python
airline.describe()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
df_norm = norm_func(airline.iloc[:,1:])
df_norm
```

```python
plt.figure(figsize=(10, 7))
dendgrogram = sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
df_norm['h_clusterid'] = Clusters
df_norm
```

```python
Clusters['Clusters'].value_counts()
```

```python
df_norm.groupby('Clusters').agg(['mean']).reset_index()
```

# K MEANS

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_airline = scaler.fit_transform(airline.iloc[:,1:])
```

```python
scaled_airline
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_airline)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
cluster_new = KMeans(4, random_state=42)
cluster_new.fit(scaled_airline)
```

```python
cluster_new.labels_
```

```python
airline['clusterid_new'] = cluster_new.labels_
airline
```

```python
airline.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
airline[airline['clusterid_new']==3]
```

```python
airline[airline['clusterid_new']==0]
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(airline.head())
```

```python
airline.info()
```

```python
array=airline.values
array
```

```python
airline.describe()
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=12)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([airline,cl],axis=1)
```
